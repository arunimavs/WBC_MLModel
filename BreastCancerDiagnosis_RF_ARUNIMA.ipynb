#####**PACKAGES TO BE IMPORTED**
------------------------------------

```
import numpy as np
import pandas as pd
from google.colab import drive
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

#####**IMPORT DATASET**
-----------------------------

```
#MOUNT THE GOOGLE DRIVE
drive.mount('/content/drive/')

#SET THE PATH AND READ THE DATASET
data = pd.read_csv('/content/drive/MyDrive/BreastCancerWinsconsin_Dataset.csv')

print(data.shape)

data.head()
```

#####**DATA PRE-PROCESSING**
--------------------------------------------

```
# DROP THE LAST EMPTY COLOUMN

data = data.drop('Unnamed: 32', axis = 1)

print(data.shape)

data.head()

# DROP UNIQUE CUSTOMER ID 

data = data.drop('id', axis = 1)

print(data.shape)

data.head()
```

```
data.describe()
```

```
data['diagnosis'].value_counts()
```

######**ENCODING THE DEPENDENT VARIABLE TO TRAIN**
--------------------------------------------------------

```
# LABEL ENCODING FOR OUR DEPENDENT VARIABLE DIAGNOSIS
le = LabelEncoder()
#TRANSFORM B AND M VALUES TO ENCODED LABELS
data['diagnosis'] = le.fit_transform(data['diagnosis'])
data['diagnosis'].value_counts()
```

#####**SET DEPENDENT AND INDEPENDENT VARIABLES**
--------------------------------------------------------

```
# GET X (independent) WHICH IS ALL THE 30 COLOUMNX EXCEPT DIAGNOSIS 
x = data.iloc[:,1:]
# AND Y (dependent) VARIABLE WHICH IS DIAGNOSIS
y = data.iloc[:,0]

print(x.shape)
print(y.shape)
```

#####**SPLIT THE DATASET USING TRAIN_TEST_SPLIT**
-----------------------------------

```
#Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 16)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

#####**MODELLING - RANDOM FOREST**
----------------------------------------

```
# creating a model
model = RandomForestClassifier(n_estimators = 400, max_depth = 10)

# feeding the training set into the model
model.fit(x_train, y_train)

# predicting the test set results
y_pred = model.predict(x_test)

# Calculating the accuracies
print("Training accuracy :", model.score(x_train, y_train))
print("Testing accuarcy :", model.score(x_test, y_test))

# classification report
cr = classification_report(y_test, y_pred)
print(cr)

# confusion matrix 
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
